```python
import numpy as np
from PIL import Image

def convert_bmp_to_raw(bmp_path, save_path, save_name, quality_factor=1024, image_format=4):
    # Load the BMP image
    image = Image.open(bmp_path)
    width, height = image.size
    image_array = np.array(image)
    
    # Flatten the image array and save as .raw file
    raw_image_data = image_array.flatten()
    raw_image_data.tofile(save_path+save_name+'.raw')
    
    # Create the accompanying info text
    info_text = f"{save_name}.raw\n{quality_factor}\n{image_format}\n{width}\n{height}\n"
    
    # Save the info text to a file
    with open(save_path+save_name+"_param.txt", 'w') as info_file:
        info_file.write(info_text)

    print(f"Conversion complete! Files saved as {save_path+save_name}.raw and {save_path+save_name}_param.txt")

# Example usage
bmp_path = './files/chameleon.bmp'  # Replace with the path to your BMP file
save_path = './files/'  # Replace with the path to the directory where you want to save the files
save_name = 'chameleon'  # Replace with the desired name of the output files

convert_bmp_to_raw(bmp_path, save_path, save_name)
```
